## In this Notebook, I will merge and structure Pollutant and Meteorological datasets, in order to get ready for being used as definitive dataset to be modelled under Machine Learning

# 1. Import libraries

```python
import numpy as np
import pandas as pd
```

# 2. Import Pollutant dataset

```python
df1 = pd.read_csv("Data_sets/Pollutant_ready.csv", index_col=0)
pd.options.display.max_columns = None
df1
```

# 3. Import Meteorological dataset

```python
df2 = pd.read_csv("Data_sets/Meteo_ready.csv", index_col=0)
pd.options.display.max_columns = None
df2
```

# 4. Merge Pollutant and Meteorological datasets

```python
#As pollutant dataset has fewer rows than meteorological one, then, datasets are merged with pollutant primary domain data 
merged = pd.merge(left=df1, right=df2, how='left', left_on='data', right_on='data_lectura')
merged
```

```python
#Let's check type features
merged.info()
```

# 5. Export Dataset merged

```python
#dataset merged as csv file
merged.to_csv(r'Data_sets/Pollutant_Meteo_merged_ready.csv')
```
